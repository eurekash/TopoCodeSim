```python
import matplotlib.pyplot as plt
from matplotlib import rc
from scipy.optimize import curve_fit
import numpy as np
```

```python

def func(x,A,B,C,D):
    return ((A*x+B)*x+C)*x+D

def sigmoid(x, L, x0, k, b):
    return L / (1 + np.exp(-k*(x-x0)))+b
    
def plot(ax, phys, err, label):
    ax.scatter(phys, err)
    phys_smooth = np.linspace(min(phys), max(phys), 1000)
    popt, _ = curve_fit(sigmoid, phys, err)
    ax.plot(phys_smooth, sigmoid(phys_smooth, *popt), label=label)
```
